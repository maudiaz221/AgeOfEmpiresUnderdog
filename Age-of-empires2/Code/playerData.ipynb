```python


def distance(s1, s2, cont):
    if len(s1) == 0 or len(s2) == 0:
        return len(s1) + len(s2) + cont
    if s1[0] == s2[0]:
        return distance(s1[1:],s2[1:],cont)
    
    sus = distance(s1[1:],s2[1:],cont+1)
    ins = distance(s1[1:],s2,cont+1)
    bor = distance(s1,s2[1:],cont+1)
    
    return min(sus,ins,bor)

distance("Alex", "Z40", 0)
    
```
