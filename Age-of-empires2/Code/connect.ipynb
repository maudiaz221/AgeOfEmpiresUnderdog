```python
import pandas as pd

# List to store each DataFrame
dataframes = []

# Loop to read each CSV and store the DataFrame in the list
for i in range(1, 13):
    df = pd.read_csv(f'players/p{i}.csv')
    dataframes.append(df)

# Concatenate all DataFrames in the list along rows (axis=0)
players = pd.concat(dataframes, axis=0)

# Optionally reset the index if needed
players.reset_index(drop=True, inplace=True)

# Display the combined DataFrame
players
```

```python
players['Born'] = players['Born'].replace({None:''})
```

```python
players
```

```python
import re

# Function to extract age from the 'born' column
def extract_age(row):
    #row = str(row)
    match = str(re.findall(r'(age\xa0\d{2})', row))
    match = re.findall(r'0\d{2}', match)
    if match:
        return match[0][1:]
    return None



# Apply the function to the 'born' column and create a new 'age' column
players['age'] = players['Born'].apply(extract_age)

#Display the DataFrame
players
```

```python
import json
with open('result.json', 'r') as file:
    data = json.load(file)

```

```python
players['User'] = players['User'].map(data)
```

```python
players.drop(columns=['Born','Tournament(AoE2)','RMVoobly(AoE2)','Alternate IDs','Games'], inplace=True)
```
