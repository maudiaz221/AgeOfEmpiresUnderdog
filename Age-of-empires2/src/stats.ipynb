# Split into different types of tournaments

```python
import re

# Function to compile regex pattern for a specific tournament type
def compile_pattern(tournament_list):
    pattern = '|'.join(map(re.escape, tournament_list))
    return re.compile(pattern, re.IGNORECASE)

# Function to check if tournament name matches a pattern
def check_pattern(tournament_name, regex):
    return bool(regex.search(tournament_name))

# Function to assign type of tournament
def assign_tournament_type(tournament_name, tournament_patterns):
    for tournament_type, pattern in tournament_patterns.items():
        if check_pattern(tournament_name, pattern):
            return tournament_type
    return None

# Dictionary mapping tournament types to filenames
tournament_files = {
    's-tier': 's-tier.txt',
    'a-tier': 'a-tier.txt',
    'b-tier': 'b-tier.txt',
    'c-tier': 'c-tier.txt',
    'FFA': 'FFA.txt',
    'Miscellaneous': 'Miscellaneous.txt',
    'Monthly': 'Monthly.txt',
    'Weekly': 'Weekly.txt',
    'Qualifier': 'Qualifier.txt',
    'Showmatch': 'Showmatch.txt'
}

# Compile regex patterns for each tournament type
tournament_patterns = {}
for tournament_type, filename in tournament_files.items():
    with open(f'../{filename}', 'r') as file:
        tournament_list = [line.strip() for line in file]
    tournament_patterns[tournament_type] = compile_pattern(tournament_list)

# Apply the function to assign tournament type
data['tournament_type'] = data['name'].apply(lambda x: assign_tournament_type(x, tournament_patterns))

data
```

## Underdog stats

```python

def checkifUnderdogwins(row, threshold):
    if abs(row['elo_player_1']-row['elo_player_2']) > threshold:
        if min(row['elo_player_1'],row['elo_player_2']) == row['elo_player_1']:
            return row['score_1'] > row['score_2']
        else:
            return row['score_2'] > row['score_1']
    else:
        return False

thresholds = [10,20,50,100]
def createplot(df,thresholds):
    data = []
    for threshold in thresholds:
        df['Underdog'] = df.apply(checkifUnderdogwins, axis=1, args=(threshold,))
        val = df['Underdog'].value_counts()
        false = val[False]
        true = val[True]
        arr = [true, false]
        data.append(arr)



    species = ("10", "20", "50", "100")

    x = np.arange(len(species))  # the label locations
    width = 0.35  # the width of the bars
    multiplier = 0

    fig, ax = plt.subplots(layout='constrained')

    data1 = {
        "True": [data[i][0] for i in range(len(species))],
        "False": [data[i][1] for i in range(len(species))]
    }
    for attribute, measurement in data1.items():
        offset = width * multiplier
        rects = ax.bar(x + offset, measurement, width, label=attribute)
        ax.bar_label(rects, padding=3)
        multiplier += 1

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Frequency')
    ax.set_title('Frequency of True and False values for each threshold')
    ax.set_xticks(x + width, species)
    ax.legend(loc='upper left', ncols=4)

    plt.tight_layout()  # Adjust layout for better spacing
    plt.show()
    
    return data
    
```
